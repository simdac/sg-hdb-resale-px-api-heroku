```python
# Import python libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Combine all csv files data and display top results

df1 = pd.read_csv('resale-flat-prices-based-on-approval-date-1990-1999.csv')
df2 = pd.read_csv('resale-flat-prices-based-on-approval-date-2000-feb-2012.csv')
df3 = pd.read_csv('resale-flat-prices-based-on-registration-date-from-jan-2015-to-dec-2016.csv')
df4 = pd.read_csv('resale-flat-prices-based-on-registration-date-from-jan-2017-onwards.csv')
df5 = pd.read_csv('resale-flat-prices-based-on-registration-date-from-mar-2012-to-dec-2014.csv')

df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#return top 5 rows 
df.head()
```

```python
#idk lol 
df.info()
```

```python
#checking for NA values, so since flat_model and flat_type are both 0 we gotta move on hehe
df.isna().sum()
```

```python
# Display the HDB Resale price group by flat_model and view some basic statistical details
df.groupby(['flat_model'])['resale_price'].describe()
```

```python
#so as u can see the flat_model is so stupid we gotta clean it 
```

```python
#capitalise all names of flat models bc idk which dummy suka suka suddenly all upper case
df.flat_model = df.flat_model.apply(lambda x: x.upper())
```

```python
# Display the new HDB Resale price group by flat_model and view some basic statistical details
df.groupby(['flat_model'])['resale_price'].describe()
```

```python
#ok so it looks more acceptable now but dk which toot go put the fullstop beside the premium apartment so we gotta get rid of it
```

```python
#rename it bc idw to remove punctuations because theres some names with hyphens
df['flat_model']=df['flat_model'].replace({'PREMIUM APARTMENT.':'PREMIUM APARTMENT'})
```

```python
# Display the best HDB Resale price group by flat_model and view some basic statistical details
df.groupby(['flat_model'])['resale_price'].describe()
```

```python
#plot dem graphs 
sns.countplot(x="flat_model", data=df)

#define countvsflat_model graph
countvsflat_model = sns.countplot(x="flat_model", data=df)

#the x-axis will overlap so this is to rotate the ticklabels
#theyre much much more pleasant looking like this hehe
countvsflat_model.set_xticklabels(countvsflat_model.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()

#btw i dont really know what count is so idk why i did this graph
```

```python
#plotplotplot
sns.barplot(x='flat_model', y='resale_price', data=df)

#define rspfvsflat_model graph
rsfpvsflat_model= sns.barplot(x='flat_model', y='resale_price', data=df)

#same thing the x-axis will overlap so this is to rotate the ticklabels
rsfpvsflat_model.set_xticklabels(rsfpvsflat_model.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```
