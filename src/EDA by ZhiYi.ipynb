```python
# Import python libraries
import pandas as pd
import numpy as np
import seaborn as sns
import glob 
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

```python
#read resale_flat_prices.csv file
df = pd.read_csv('resale_flat_prices.csv')
```

```python
#top 5 rows 
df.head()
```

```python
#observe data types of each column
df.info()
```

```python
df.isna().sum()
```

```python
# Display the HDB Resale price group by flat_model and view some basic statistical details
df.groupby(['flat_model'])['resale_price'].describe()
```

```python
#capitalise all the names of flat model
df.flat_model = df.flat_model.apply(lambda x: x.upper())
```

```python
# Display the HDB Resale price group by flat_model and view some basic statistical details
df.groupby(['flat_model'])['resale_price'].describe()
```

```python
# replace the name because theres a fullstop 
df['flat_model']=df['flat_model'].replace({'PREMIUM APARTMENT.':'PREMIUM APARTMENT'})
```

```python
# Display the HDB Resale price group by flat_model and view some basic statistical details
df.groupby(['flat_model'])['resale_price'].describe()
```

```python
# making the graph bigger 
fig1_dimensions=(25, 8)
fig1 = plt.figure(figsize = fig1_dimensions)
ax1 = fig1.add_subplot(1,3,1)

#plot graph of count vs flat model
sns.countplot(x="flat_model", data=df)
countvsflat_model = sns.countplot(x="flat_model", data=df)

#the x-axis will overlap so rotate the ticklabels
countvsflat_model.set_xticklabels(countvsflat_model.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

### The most common type of flat is Model A, followed by Improved. 

```python
#plot graph of rsfp vs flat model
sns.barplot(x='flat_model', y='resale_price', data=df)
rsfpvsflat_model= sns.barplot(x='flat_model', y='resale_price', data=df)

#rotate ticklabel
rsfpvsflat_model.set_xticklabels(rsfpvsflat_model.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

### Type S2 has the highest resale price whereas the 2-Room flat model have the lowest resale price

```python
# Display the HDB Resale price group by flat_type and view some basic statistical details
df.groupby(['flat_type'])['resale_price'].describe()
```

```python
#replace multi-generation with multi generation
df['flat_type']=df['flat_type'].replace({'MULTI-GENERATION':'MULTI GENERATION'})
```

```python
# Display the HDB Resale price group by flat_type and view some basic statistical details
df.groupby(['flat_type'])['resale_price'].describe()
```

```python
#plot graph of count against flat type
sns.countplot(x="flat_type", data=df, log=True)
df = df.sort_values(['flat_type']).reset_index(drop=True)

#define countvsflat_model graph
countvsflat_type = sns.countplot(x="flat_type", data=df, log=True)

#the x-axis will overlap so this is to rotate the ticklabels
countvsflat_type.set_xticklabels(countvsflat_type.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()

fig2_dimensions=(25, 8)
fig2 = plt.figure(figsize = fig2_dimensions)
```

### The most common flat type is 4 Room, followed by 3 Room and 5 Room. The least common flat type would be Multi Generation, but it was only first rolled out in 2013 and only started gaining popularity in 2018.

```python
#plot graph of resale price against flat type
sns.barplot(x='flat_type', y='resale_price', data=df)
rsfpvsflat_type= sns.barplot(x='flat_type', y='resale_price', data=df)

#rotate ticklabel
rsfpvsflat_type.set_xticklabels(rsfpvsflat_type.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

### Resale price increases as number of rooms and size of flat type increases.

```python
#plot graph of count against flat type
sns.countplot(x="flat_type", data=df)
countvsflat_type = sns.countplot(x="flat_type", data=df)

#the x-axis will overlap so this is to rotate the ticklabels
countvsflat_type.set_xticklabels(countvsflat_type.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

```python
#creating pivot table 
pivotdf = pd.pivot_table(df[['flat_type','resale_price','town']], 
               index=['town'], 
               columns=['flat_type'],
               values=['resale_price'],
               aggfunc={"resale_price":[np.mean]}).fillna(0)
pivotdf.reset_index(inplace=True)
pivotdf.columns =['town','1 ROOM','2 ROOM','3 ROOM','4 ROOM','5 ROOM','EXECUTIVE','MULTI GENERATION']

# Plot the stacked bar chart here
pivotdf = pivotdf[['town', '1 ROOM', '2 ROOM', '3 ROOM', '4 ROOM', '5 ROOM', 'EXECUTIVE', 'MULTI GENERATION']]

fig1_dimensions=(16, 10)
fig1 = plt.figure(figsize = fig1_dimensions)
ax1 = fig1.add_subplot(1,1,1)
pivotdf.plot(kind='bar',
                    x='town',
                    stacked=True,  # <<< stacked
                    figsize=(16,8), 
                    title="Resale Price by town",
                    ax=ax1)
ax1.set_ylabel("Resale_Price")

plt.show()
```
